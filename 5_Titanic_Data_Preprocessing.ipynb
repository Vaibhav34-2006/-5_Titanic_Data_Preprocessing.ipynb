```
#### Data Wrangling (Working With Null Values)
```

```
titanic_data.info()
```

```
titanic_data.isnull().sum()
```

```
mean_age = titanic_data['Age'].mean()
mean_age = np.round(mean_age,2)
mean_age
```

```
titanic_data.fillna(mean_age,inplace=True)
```

```
titanic_data.isnull().sum()
```

```
titanic_data.head()
```

```
###titanic.dropna(inplace=True)
```

```
pclass = pd.get_dummies(titanic_data['Pclass'],drop_first=True)
pclass.head(10)
```

```
sex = pd.get_dummies(titanic_data['Sex'],drop_first=True)
sex.head(10)
```

```
titanic_data = pd.concat([titanic_data,sex,pclass],axis=1)
titanic_data.head()
```

```
titanic_data.drop(["Sex","Pclass"],axis=1,inplace=True)
```

```
titanic_data.head()
```

```
titanic_data.rename(columns={"male":"sex",2:"pclass_2",3:"pclass_3"},inplace=True)
```
